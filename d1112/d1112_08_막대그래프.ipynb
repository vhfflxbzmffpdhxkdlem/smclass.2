```python
import matplotlib.pyplot as plt
# 한글성정 
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
```

### 라벨 변경,라벨의 기울기 설정, width,alpha 투명도 조정

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]
# color : 색상, width :  바크기조절
colors=['r','g','b']
plt.figure(figsize=(5,3))   # 그래프 크기 조절
plt.bar(label,values,color=colors,width=0.3,label='학생키')  # 바색상을 여러개로 처리
plt.yticks([0,50,100,150,200,250])
plt.xticks(label,['1번학생','2번학생','3번학생'],rotation=45)  # x라벨 변경
plt.title('막대그래프')  # 상단제목
plt.legend(loc='upper center')

# plt.bar(label,values,color='r')  # 1개의 색상으로 처리
# plt.plot(label,values)  # 선그래프
```

막대그래프 무늬설정

```python
colors = ['red','green','blue']
bar = plt.bar(label,values,width=0.5,color=colors,alpha=0.5)
# 막대그래프의 무늬
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

* 0부터 범위설정 xticks, yticks = 0부터 표시됨
* 특정구간부터 범위설정 : xlim, ylim

```python
colors = ['red','green','blue']
bar = plt.bar(label,values,width=0.5,color=colors,alpha=0.5)
# plt.yticks([0,100,150,200])  # 0부터 범위지정
plt.ylim(175,190) # 특정구간 범위지정
plt.show()
```

barh : 옆으로 누운 막대그래프

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]
colors = ['red','green','blue']
plt.barh(label,values,height=0.5,color=colors)
plt.xlim(175,190)
```

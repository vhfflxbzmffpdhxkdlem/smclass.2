```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 한글성정 
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용

df=pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# 복합그래프 - 여러개 그래프
## 2x2 = 4개 그래프를 혼합해서 사용
fig,axs = plt.subplots(2,2,figsize=(15,13))
# 메인타이틀 생성
fig.suptitle('4개 복합 그래프')

# 4개 그래프 : 1번째 axs[0,0],[0,1],[1,0],[1,1]
## set명령어 사용
## 1번째 bar 그래프 입력
axs[0,0].bar(df['이름'],df['국어'],label='국어')
axs[0,0].set_title('1. 국어점수 그래프')
# x축,y축 이름
axs[0,0].set(xlabel='이름',ylabel='점수')
# 배경색
axs[0,0].set_facecolor('lightyellow')
# 범례
axs[0,0].legend(loc='upper center')

## 2번째 : 선그래프
axs[0,1].set_title('2. 영수,수학점수 그래프')
axs[0,1].set(xlabel='이름',ylabel='점수')
axs[0,1].plot(df['이름'],df['영어'],label='영어')
axs[0,1].plot(df['이름'],df['수학'],label='수학')
axs[0,1].legend()

## 3번째 :
axs[1,0].set_title('3. 가로바 키 그래프')
axs[1,0].set(xlabel='이름',ylabel='점수')
axs[1,0].set_xlim(150,210)
axs[1,0].barh(df['이름'],df['키'],label='키')
for i in range(8):
  axs[1,0].text(df['키'][i]+2,df['이름'][i],df['키'][i],ha='center')
axs[1,0].legend()

## 4번째 : 선그래프
axs[1,1].set_title('4. 사회점수 선 그래프')
axs[1,1].plot(df['이름'],df['사회'],label='사회',color='g',alpha=0.5,marker='o')
for i in range(8):
  axs[1,1].text(df['이름'][i],df['사회'][i]+2,df['사회'][i],ha='center')
axs[1,1].legend()
```

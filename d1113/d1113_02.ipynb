```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 한글성정 
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
def cm_func(x):
  return str(x)+'cm'

# df['키'] : 배열타입의 모든 것을 함수 적용 - apply
df['키'] = df['키'].apply(cm_func)
df['키']
```

```python
## 50점 아래 있는 값 : 50으로 리턴해서 출려하시오.

def k_func(x):
  if x>=50: return x
  return 50

### int,float,str 단일값은 함수명(데이터)으로 호출
for i in range(8):
  print(k_func(df['국어'][i]))
```

```python
def h_func(x):
  return x/2
### 누적막대그래프를 사용해서 값들을 모두 중앙에 출력하시오.


plt.bar(df['이름'],df['국어'])
plt.bar(df['이름'],df['영어'],bottom=df['국어'])
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'])
## 데이터가 단일타입 : int,float.str
for i in range(8):
  plt.text(df['이름'][i],h_func(df['국어'][i]),df['국어'][i],ha='center')
  plt.text(df['이름'][i],df['국어'][i]+h_func(df['영어'][i]),df['영어'][i],ha='center')
  plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]+h_func(df['수학'][i]),df['수학'][i],ha='center')

plt.show()
```

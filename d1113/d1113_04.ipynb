```python
# 원그래프
values = [30,25,20,13,10,2]   # 100퍼센트
# values = [1,1,1,1,2,1]
explode = [0.1,0.1,0,0,0,0]
colors = ['plum','beige','skyblue','bisque','khaki','peachpuff']
labels = ['python','java','javascript','oracle','html','etc']
## edgecolor : 테두리  linewidth : 테두리 두께
wedgeprops = {'width':0.6,'edgecolor':'white','linewidth':2}

### 데이터 값이 : 2% 이하의 값은 출력이 되지 않도록 함수 적용
### pct : values의 퍼선트 값 생성
# def custom_autopct(pct):
  # if pct >= 10: 
  #   # return ('%.1f%%' % pct)
  #   #return ('{:.1f}%'.format(pct))
  #   return(f'{pct:.1f}%')
  # return ''
def custom_autopct(pct): 
  return f'{pct:.1f}%' if pct >=10 else f'{pct:.0f}%'

## wedgeprops : 중간공백 ,pctdistance: 글자위치
plt.pie(values,labels=labels,colors=colors,startangle=90,counterclock=False,\
        autopct=custom_autopct,wedgeprops=wedgeprops,pctdistance=0.7)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 한글성정 
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
n_list = [0,1,2,3,4,5,6,7]
```

```python
plt.title('학생성적 그래프')
plt.plot(df.index,df['국어'],label='국어',marker='x',ls=':',ms=10)
plt.plot(df.index,df['영어'],label='영어',marker='v',ls='--',ms=10)
plt.plot(df.index,df['수학'],label='수학',marker='o',ls='-.',ms=10)
for i in range(8):
  plt.text(df.index[i],df['국어'][i]+1,df['국어'][i],ha='center')
  plt.text(df.index[i],df['영어'][i]+1,df['영어'][i],ha='center')
  plt.text(df.index[i],df['수학'][i]+1,df['수학'][i],ha='center')
plt.legend()
plt.show()
```

```python
def func(x):
  if x-30<0: return 0
  else: return x-30
```

```python
def n_func(x):
  if x-50>0: return x
  else: return 0

# apply : pandas 사용
df['국어'].apply(n_func)
```

```python
# 누적막대그래프
plt.title('학생성적그래프')
plt.bar(df['이름'],df['국어'],label='국어',width=0.6,color='r')
plt.bar(df['이름'],df['영어'],label='영어',width=0.6,bottom=df['국어'],color='y')
plt.bar(df['이름'],df['수학'],label='수학',width=0.6,bottom=df['국어']+df['영어'],color='g')
for i in range(8):
  plt.text(df['이름'][i],df['국어'][i]-12,df['국어'][i],ha='center')
  plt.text(df['이름'][i],df['국어'][i]-12+df['영어'][i],df['영어'][i],ha='center')
  plt.text(df['이름'][i],df['국어'][i]-12+df['영어'][i]+df['수학'][i],df['수학'][i],ha='center')
plt.ylim(0,350)
plt.legend(ncol=3)
n_list = df['국어'].apply(func)
plt.figure(figsize=(10,4))
plt.show()
```

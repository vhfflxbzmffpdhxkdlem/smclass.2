```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 한글성정 
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
```

```python
# 남자인구현황 그래프
df=pd.read_excel('연령별인구현황_202410.xlsx',skiprows=3,usecols='B,E:Y',nrows=1,index_col='행정기관')
# 여자인구현황 그래프
df2=pd.read_excel('연령별인구현황_202410.xlsx',skiprows=3,usecols='B,AB:AV',nrows=1,index_col='행정기관')
```

```python
# 연령별 막대그래프로 출력하시오
## 1. 데이터값, 천단위 표시 제거, int타입 변경
## 2. plt.bat(x,y)출력
```

```python
df.iloc[0] = df.iloc[0].str.replace(",","").astype(int)
df.rename(index={'전국  ':'전국'},inplace=True)

df2.iloc[0] = df2.iloc[0].str.replace(",","").astype(int)
df2.rename(index={'전국  ':'전국'},inplace=True)
```

```python
plt.barh(df.columns,df.loc['전국'])
plt.title('2024년 인구수 남')
plt.show()

plt.barh(df2.columns,df2.loc['전국'])
plt.title('2024년 인구수 여')
plt.show()
```

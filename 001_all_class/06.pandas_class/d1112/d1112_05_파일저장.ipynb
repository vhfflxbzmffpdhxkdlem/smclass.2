```python
import pandas as pd
import random
df = pd.read_excel("score.xlsx",index_col='지원번호')
```

```python
import matplotlib.pyplot as plt
# 한글성정 
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
```

```python
x = df.index
k = df['국어']
e = df['영어']
# 그래프 크기 조절, dpi : 그래프해상도 설정  해상도가 높을수록 이미지 크기가 커짐.
# facecolor : 배경색 설정
plt.figure(figsize=(20,5),facecolor='yellow') # x축으로 5,y축으로는 20 크기조절
# plt.figure(figsize=(20,5),dpi=1000) # x축으로 5,y축으로는 20 크기조절
plt.plot(x,k,label='국어',marker='o',ms=15,color='red',mfc='yellow',mec='blue',ls='--')
plt.plot(x,e,label='영어',marker='v',ms=15)
plt.legend()

# 그래프 파일로 저장
plt.savefig('graph,png2',dpi=300)
```

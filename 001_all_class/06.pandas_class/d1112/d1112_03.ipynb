```python
import matplotlib.pyplot as plt
# 한글성정 
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #Win용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
```

```python
## x,y축의 데이터
x = [-1,2,3]   # 리스트타입,튜플,Dataframe데이터 - 리스트형태,Series
y = [-2,-4,8]   
## 선그래프 생성
plt.plot(x,y)
# ## 상단타이틀
plt.title("라인그래프 Line",size = 50)  # size : 글자크기 변경
# 그래프X축 설명, color : 글자색변경/색상값,16코드, loc : 글자위치 left,center,right
plt.xlabel('X축 날짜',color = '#1485b5',loc='center')  
# 그래프Y축 설명, color : 글자색변경/색상값,16코드, loc : 글자위치 top,center,bottom
plt.ylabel('Y축 인원',color = 'blue',loc='center')  

## 그래프의 눈금단위 표시
plt.xticks([1,2,3])
plt.yticks([2,4,10])
plt.show()
```
